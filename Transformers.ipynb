```python
# Step1: specify our feature extraction model
feature_extractor = pipeline("feature-extraction",
                             model="sentence-transformers/all-MiniLM-L6-v2")


## step2: create a movies x embeddings array (166, 384)
embeddings = movies.movie_info.apply(lambda row: feature_extractor(row)[0][0])
embddings_movies = np.vstack(embeddings)


## step3: create a get similar movies function with the inputs: embeddings, movie index, movie_title, movie_info, top_n

def get_similar_movies(embeddings, movie_index, movie_details, top_n=3):
  # create movie embedding for movie_index
    movie_embedding = np.array(embeddings[movie_index]).reshape(1, -1)
  # calculate similarity scores
    similarity_scores_movie = cs(movie_embedding, embeddings)
    similarity_scores_movie_series = pd.Series(similarity_scores_movie.flatten(), name='similarity_score')
  # bring in movie info
    movies_similarity_scores_df = pd.concat([movie_details, similarity_scores_movie_series], axis=1)
  # display movie recs
    return movies_similarity_scores_df.sort_values('similarity_score', ascending=False).iloc[0:top_n + 1]
```

```python
get_similar_movies(embddings_movies, 25, movies[['movie_title', 'movie_info']])
```

```python
get_similar_movies(embddings_movies, 131, movies[['movie_title', 'movie_info', 'rating']], top_n=5)
```
