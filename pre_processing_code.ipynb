## NLP Pipeline

```python
def lower_replace(series):
  return series.str.lower().str.replace(r'\[.*?\]', '', regex=True).str.replace(r'[^\w\s]', '', regex=True)

def token_lemma_nonstop(text):
  return ' '.join([token.lemma_ for token in nlp(text) if not token.is_stop])

def filter_pos(text, pos_list = ['NOUN', 'PROPN']):
  return ' '.join([token.text for token in nlp(text) if token.pos_ in pos_list])
```

```python
def nlp_pipeline(series):
  return lower_replace(series).apply(token_lemma_nonstop).apply(filter_pos)
```

```python
cv = CountVectorizer(stop_words='english', ngram_range=(1,2), min_df=2)
dtm = cv.fit_transform(series)
data_dtf = pd.DataFrame(dtm.toarray(), columns=cv.get_feature_names_out())
data_dtf
```
